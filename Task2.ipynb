```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
data = pd.read_csv(r'C:\Users\Shanu\Desktop\Elevate Task\Titanic-Dataset.csv')#this read data which is located at specific path
print(data)
```

```python
data.describe()
```

```python
data.median(numeric_only=True)#this numeric_only=True exclude str part take only column that contain only numeric value
```

```python
numeric_cols = data.select_dtypes(include=['int64', 'float64']).columns# this part is also done in task 1 we take only columns that has numeric features and store it in list
print("Numeric columns:", list(numeric_cols))
```

```python
# creating histogram chart
for i in numeric_cols:
    plt.figure(figsize=(5, 3))#tell fig size
    plt.hist(data[i].dropna(),color='skyblue', edgecolor='black')# in this histogram first select the col then remove missing value for reducing error then color of histo vale is ble and fig edge blacked
    plt.title(f'Histogram of {i}') # give title of every col that are looped in i
    plt.xlabel(i) #x axis contain our Numeric columns
    plt.ylabel('Frequency') #it contain frequency
    plt.show()
```

```python
# creating boxplot chart
for j in numeric_cols:
    plt.figure(figsize=(5, 3))
    sns.boxplot(data[j])
    plt.title(f'Boxplot of {j}')
    plt.ylabel(j)
    plt.show()
```

```python
# now doing 3rd point 
cor_matrix = data.corr(numeric_only=True)
# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(cor_matrix,)
plt.title("Correlation Matrix")
plt.show()
```

```python
#now doing pairplot
sns.pairplot(data)
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```
